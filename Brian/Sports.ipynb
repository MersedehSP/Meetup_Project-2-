```python
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
executable_path={'executable_path':ChromeDriverManager().install()}
browser=Browser('chrome',**executable_path, headless=False)
```

```python
#url='https://www.meetup.com/find/us--pa--philadelphia/sports-fitness/?eventType=inPerson'
url='https://www.meetup.com'
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
#soup.prettify()
```

```python
abc = soup.find_all('div', class_= 'css-13zb9hu')

categories = []
for x in abc:
    div_each_category         = x.find('a', class_='css-1i2on81')
    div_each_category_a_div   = div_each_category.find('div', class_='css-46hqvx')
    
    div_each_category_a_div_2 = div_each_category_a_div.find('div', class_='css-qpi38e')
    
    categorytext              = div_each_category_a_div_2.text
    categories.append(categorytext)
    
    

print(categories)
```
