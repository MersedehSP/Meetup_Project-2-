```python
#import dependencies
import pandas as pd
import numpy as np
```

```python
# Read venue data from events_venue.csv

venue_df = pd.read_csv('Final_Events.csv')
venue_df
```

```python

venue_df = venue_df.rename(columns={"event_loc_state": "event_state", "event_loc_city": "event_city"})
venue_df
```

```python


# Replace all the empty cells with NaN
venue_df = venue_df[:].replace(' ', np.NaN)

# Convert the event_state column data to uppercase
venue_df['event_state']= venue_df['event_state'].str.upper()
venue_df.head(50)
```

```python
# Read all the city events data from Final_data.csv

events_df = pd.read_csv('Final_data.csv')
events_df
```

```python
# Rename the event_link column in the venue data frame

venue_df = venue_df.rename(columns={"event_link": "venue_event_link"})
venue_df
```

```python
# Concatenate the venue dataframe to the city dataframe on column

final_df = pd.concat([events_df, venue_df], axis=1)
final_df.head(50)
```

```python
# Drop venue_event_link column since it is the same as event_link

new_df = final_df.drop(final_df.columns[6], axis=1)

new_df = new_df.dropna(how='any')
new_df.head(50)
```

```python
# Export the df to a csv file

new_df.to_csv('Final_Meetup_Events.csv')
```
