```python
import json
import csv
import pprint
```

```python
csvfilepath = "complete_events_lat_lng.csv"
jsonfilepath = "jsonifydata.json"
```

```python
data = []
```

```python
with open(csvfilepath,encoding="utf8") as csvfile:
    csvReader = csv.DictReader(csvfile)
    for rows in csvReader:
        # id = rows['id']
        # data[id] = rows
        data.append(rows)
```

```python
with open(jsonfilepath,"w") as jsonFile:
    jsonFile.write(json.dumps(data,indent=4))
```

```python
pprint.pprint(data)
```
